**Obs:** Os elementos tr e td são tags HTML que são comumente usadas para criar tabelas.

No exemplo abaixo existem duas linhas que devem de ser excluídas, por isso o se não ter td: continue

```python
table = soup.find_all("table")[0]

rows = []
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    if not tds:
        continue
    row = [td.text.strip() for td in tds]
    rows.append(row)

rows = rows[2:]
```

```python
#Convertendo as cooordenadas em padrões interessantes
def convert_coord(coord):
    parsed_coord = coord.split()
    degree, minutes, seconds = [int(i) for i in parsed_coord[1::2]]
    final_degree = degree + minutes / 60 + seconds / (60*60)
    return final_degree
```

**Considerações:**

O código Python define uma função chamada convert_coord que aceita uma coordenada como argumento. A função converte uma coordenada em formato de graus, minutos e segundos para um valor em graus decimais. Vamos entender cada parte da função:

A entrada coord é uma string que contém graus, minutos e segundos, separados por espaços. Por exemplo, a entrada pode ser algo como "D 32 M 18 S 23".
Etapas do Código
parsed_coord = coord.split():

Este comando divide a string coord em uma lista de substrings usando o espaço como delimitador. No exemplo dado, a lista resultante seria ['D', '32', 'M', '18', 'S', '23'].
degree, minutes, seconds = [int(i) for i in parsed_coord[1::2]]:

Aqui, a lista parsed_coord é dividida em duas: uma para os caracteres alfabéticos ('D', 'M', 'S') e outra para os números (32, 18, 23).
O slicing parsed_coord[1::2] pega os elementos de índice ímpar da lista, ou seja, ['32', '18', '23'].
Esses elementos são então convertidos em inteiros e armazenados nas variáveis degree, minutes e seconds.
final_degree = degree + minutes / 60 + seconds / (60*60):

Este é o cálculo para converter graus, minutos e segundos em graus decimais.

```python
parsed_rows = []
for row in rows:
    try:
        new_row = []
        new_row.append(row[0])
        new_row.append(row[1])
        new_row.append(row[2])
        new_row.append(-1 * convert_coord(row[3]))
        new_row.append(-1 * convert_coord(row[4]))
        parsed_rows.append(new_row)
    except ValueError:
        pass
        
print(parsed_rows)
```

An empty list called parsed_rows is initialized.
The code iterates over each row in rows.
Inside the loop, it tries to create a new_row list that takes the first three elements of the current row and adds them to new_row.
It then takes the 4th and 5th elements of row, applies a convert_coord function to them, multiplies the result by -1, and adds these new elements to new_row.
Finally, new_row is appended to parsed_rows.
There are some things to note:

The try block will catch ValueError exceptions but will silently pass them, meaning that if a ValueError occurs, the code will continue without doing anything.
The code assumes that convert_coord is a function that has been previously defined. If it's not, this code will throw a NameError.
The code also assumes that each row has at least 5 elements. If it doesn't, attempting to access an index that does not exist will throw an IndexError
